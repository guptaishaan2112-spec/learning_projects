```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
sns.get_dataset_names()
df=sns.load_dataset('penguins')
print(df)
```

```python
graphs=sns.PairGrid(data=df,hue='sex',palette='Set2')
graphs.map_upper(sns.scatterplot)
graphs.map_lower(sns.kdeplot)
graphs.map_diag(sns.kdeplot,fill=True)
```

```python
sns.pairplot(data=df,diag_kind='kde',hue='species')
```

```python
sns.rugplot(data=df, y='body_mass_g',hue='species',height=0.5)
```

```python
sns.rugplot(data=df, x='body_mass_g',hue='species',height=0.5)
```

```python
columns=['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']
sns.heatmap(data=df[columns].corr(),annot=True,linewidth=2,linecolor='black',cmap='plasma')
```

```python
sns.kdeplot(data=df, x='body_mass_g',hue='island',fill=True)
```

```python
sns.violinplot(data=df, x='species', y='body_mass_g',inner=None)
sns.swarmplot(data=df,x='species',y='body_mass_g',color='red')
#the swarmplot is totally inside vioin plot
```

```python
sns.boxplot(data=df,x='species',y='body_mass_g',hue='island')
```

```python
sns.countplot(data=df,x='species')
```

```python
sns.barplot(data=df,x='species',y='body_mass_g',hue='island',palette='Set2',estimator=np.mean)
```

```python
sns.set_style('whitegrid')
sns.jointplot(data=df,x='body_mass_g', y='flipper_length_mm',hue='sex', palette=['blue','green']
              #,kind=''  -->this will change the type of graph
              )
sns.despine()
```

```python
sns.lineplot(data=df,x='body_mass_g', y='flipper_length_mm',hue='island',style='sex',palette=['red','blue','green'])
```

```python
sns.regplot(data=df, x='body_mass_g', y='flipper_length_mm',color='green',
            #scatter=False    --> it will merge all the dots into the regression line
            )
```

```python
sns.histplot(data=df,x='body_mass_g',hue='sex',multiple='layer')
```

```python
sns.swarmplot(data=df,x='species',y='body_mass_g',hue='island',palette='Set2')
```

```python
sns.stripplot(data=df,x='species',y='body_mass_g',hue='island',dodge=True,palette='Set2',alpha=0.7,jitter=True)
```

```python
sns.set_style('ticks')
sns.scatterplot(data=df,x='flipper_length_mm',y='body_mass_g',hue='island',palette='Set2',style='sex',alpha=0.7)
sns.set_context('notebook')
#sns.despine(left=True)
```
