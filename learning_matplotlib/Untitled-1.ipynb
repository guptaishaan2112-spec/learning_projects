```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
plt.style.use('default')#you may also use seaborn or ggplot in place of default
df=pd.read_csv('sales.csv')
```

```python
size=[25,30,25,20]
cat=['A','B','C','D']
c=['red','blue','pink','magenta']
plt.pie(size,labels=cat,shadow=True,colors=c,explode=(0.1,0,0,0),startangle=90,autopct='%1.1f%%')
plt.savefig('pie.png',bbox_inches='tight')
plt.savefig('pie.pdf',bbox_inches='tight')
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
max_sin=np.argmax(y)
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(x,y,label='sin wave')
ax.annotate('maximum',
            xy=(x[max_sin],y[max_sin]),
            xytext=(x[max_sin],y[max_sin]+0.3),
            arrowprops=dict(arrowstyle='->',lw=2,color='red')
)
ax.text(0.5,0.95,'this is sin wave',
        transform=ax.transAxes,
        fontsize=12,
        bbox=dict(facecolor='yellow', alpha=0.5),
        verticalalignment='top'
)
ax.axvline(x=5,color='green',linestyle='--',alpha=0.3,label='x=5')
ax.axhline(y=0,color='green',linestyle='-')
ax.legend()
ax.set_title('annotated plot')
ax.grid(True,alpha=0.3)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(8,4))

ax.plot(x,y)

ax.text(
    5,
    0,
    'Data (5,0)',
    fontsize=12,
    bbox=dict(facecolor='yellow', alpha=0.5)
)

ax.set_title('Text using Data Coordinates')
ax.grid(True)

plt.show()
```

```python
fig, axes= plt.subplots(2,1,figsize=(10,6))
df.plot(x='OrderDate',y='Sales',ax=axes[0],color='blue')
axes[0].set_xlabel('date')
axes[0].set_ylabel('value')
a=df.groupby('Product')['Sales'].sum()
a.plot(kind='bar',ax=axes[1],color='blue',edgecolor='black')
axes[1].grid(axis='y',alpha=0.3)
plt.tight_layout()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['OrderDate'],df['Sales'],'y-',linewidth=2,label='sales')
plt.plot(df['OrderDate'],df['Profit'],'r-',linewidth=2,label='profit')
plt.xlabel('date')
plt.ylabel('value')
plt.title('pandas and matplotlib')
plt.grid(True,alpha=0.3)
plt.xticks(rotation=45)
plt.legend()
```

```python
x=np.linspace(0,10,100)
fig, axes =plt.subplots(2,1,figsize=(10,6))
axes[0].plot(x,np.sin(x),'b-',linewidth=2)
axes[0].set_title('sin wave')
axes[0].grid(True,alpha=0.3)
np.random.seed(42)
a=np.random.randn(100)
b=np.random.randn(100)
scatter=axes[1].scatter(a,b,c=a,cmap='plasma')
fig.colorbar(scatter,ax=axes[1])
axes[1].set_title('scatter')
plt.tight_layout()
plt.show()
```

```python
np.random.seed(42)
data1=np.random.normal(100,15,1000)
data2=np.random.normal(120,20,1000)
plt.figure(figsize=(10,6))
plt.hist(data1,bins=30,color='skyblue',alpha=0.6,label='data1')#edgecolor and linewidth can also be added
plt.hist(data2,bins=30,color='pink',alpha=0.6,label='data2')
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('histogram of uniform distribution')
plt.grid(True,alpha=0.3,axis='y')
plt.legend()
```

```python
categories=['Q1','Q2','Q3']
product_a=[25,30,35]
product_b=[30,25,20]
x=np.arange(len(categories))
width=0.25
plt.bar(x-width/2,product_a,width,color='blue',label='product a',edgecolor='black',linewidth=1.2)
plt.bar(x+width/2,product_b,width,color='red',label='product b',edgecolor='black',linewidth=1.2)
plt.xlabel('quarter')
plt.ylabel('sales')
plt.title('sales in quarter')
plt.xticks(x,categories)
plt.grid(True,alpha=0.3)
plt.legend()
```

```python
np.random.seed(42)
x=np.random.randn(100)
y=np.random.randn(100)
size=np.random.rand(100)*1000#random.rand generate random no. less than 1 thats why *1000
color=np.random.rand(100)
plt.figure(figsize=(10,6))
plt.scatter(x,y,alpha=0.7,label='scattered',c=color,s=size,cmap='plasma',marker='o')
plt.colorbar(label='color intensity')
plt.grid(True,alpha=0.3)
plt.legend()
```

```python
x=np.linspace(-10,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,z,label='cos(x)')
plt.plot(x,y,label='sin(x)')
plt.xlabel('X axis')
plt.ylabel('sin(x) ans cos(x)')
plt.title('sin and cos graph')
plt.legend()
plt.grid(True)
plt.show()
```
